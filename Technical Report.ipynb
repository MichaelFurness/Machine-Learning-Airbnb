# Model on non-log transformed data

Following the EDA we saw that price, our predictor varaible, needed to be log transformed due to its distribution. To ensure that this was the case, I decided to run a simple linear regression model on the original price data. I would use this model as a base line score that I would need to improve on.

### Linear Regression Model

```python
final_df = pd.read_csv('../Non_log AirBnB' )
final_df.pop('Unnamed: 0')
```

```python
#assign the X and y varaibles
y = final_df.pop('price')
X =final_df
```

```python


#do a train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)

# standadise the data
scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns) 

#fit the model and print the scores
lin_model = LinearRegression()

lin_model.fit(X_train, y_train)

y_hat_1 = lin_model.predict(X_test)

print("Training Score:", lin_model.score(X_train, y_train))
print('Cross Val Scores:', cross_val_score(lin_model, X_train, y_train, cv=5))
print('Cross Val Mean Score:', cross_val_score(lin_model, X_train, y_train, cv=5).mean())
print('Test Score:', lin_model.score(X_test, y_test))
print('MSE:', mean_squared_error(y_test, y_hat_1))

# get the coefficents
# collect the model coefficients in a dataframe
df_coef = pd.DataFrame(lin_model.coef_, index=X_train.columns,
                       columns=['coefficients'])


# calculate the absolute values of the coefficients
df_coef['coef_abs'] = df_coef.coefficients.abs()
df_coef.sort_values('coef_abs', ascending=False).head(10)
```

### Result Summary

The scores for this model are poor, espically with the cross validation scores being so low, this means that our model is performing espically bad. With the distribution of the price data we would not expected high scores, this highlights the neccesaty to transform our y varaible. Due to the socres we can't interpret to much from the data as it how not performed particulay well.

Due to the models poor scores it is not neccessary to do a full visulisation of the results, however I felt it neccessary to observe the feature coefficents as it would be insightful to compare these to the log transformed models.



# Model on log transformed Price Varaible

Once I have loaded the data and removed the unwanted old index I need to scale the data and split it into a train and test split.

As the price column had been removed as it is the predictor varaible our values were of similar measure so it was not neccessary to standadise the data, however for best practice reasons I did scale my X varaibles. 

To ensure the model does not over fit and to see how it would react with unseen data I have created a train test split, with the split being 80-20.

```python
final_log_df = pd.read_csv('../log_AirBnB' )
```

```python
final_log_df.pop('Unnamed: 0')
```

```python
y_log = final_log_df.pop('price')
X_log =final_log_df
```

```python
#do a train test split
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.2, random_state=1)
```

```python
# standadise the data

scaler = StandardScaler()

X_train_log = pd.DataFrame(scaler.fit_transform(X_train_log), columns=X_log.columns)
X_test_log = pd.DataFrame(scaler.transform(X_test_log), columns=X_log.columns) 
```

```python
#function to get all the scores
def get_accuracy(model, X_train, y_train, X_test, y_test, cv=5):
    model.fit(X_train, y_train)
    scores_train = model.score(X_train, y_train)
    cr = cross_val_score(model, X_train, y_train, cv=cv).mean()

    sm = scores_train
    print("Training score: {:0.3}".format(sm))
    print("Cross Val Mean Score:", cr)
    print("Test score: {:0.3}".format(model.score(X_test, y_test)))
    return sm
```

# Linear Regression Model

To gain an understanding of the data set once modeled I decided to run a simple linear regression. This would also allow me see if the scores have improved once the data has been log transformed.

# Ada Boosting Decision Tree Regressor 

Following the improved results from the ridge and lasso models, I decided to use the versatile decison tree machline leanring algorithm as they are very powerful and capable of fitting complex datasets. Using the decison tree as the base model, I decided to use the Ensemble method of boosting which trains predictors sequentially with each model trying to correct the previous based on weaknesses. 

I decided to first use an Ada boosting ebsemble which focuses on the instances where the previous step underfitted.

```python
DT = DecisionTreeRegressor()

AB_model = AdaBoostRegressor(base_estimator=DT, random_state=1)


params = {"base_estimator__max_depth": [6,8,None],
          "base_estimator__max_features": [0.1,1,'auto'],
          "n_estimators": [100,1000],
         "loss": ['linear']}

GS_AB = GridSearchCV(AB_model, params,cv=5 ,verbose=1, return_train_score=True, n_jobs=6)

GS_AB.fit(X_train_log, y_train_log)


AB_pred = GS_AB.predict(X_test_log)

```

```python
print("Best Parameters")
GS_AB.best_estimator_
```

```python
#print the bets estimators

print("Ada Boosting Model Scores")
print("-------------------------")
print("Train R-squared:", GS_AB.score(X_train_log,y_train_log))
#print("Cross Val Score: %.2f", cross_val_score(GS_AB, X_train_log, y_train_log).mean())
print("Test Score:", GS_AB.score(X_test_log, y_test_log))
print('MSE:', mean_squared_error(y_test_log, AB_pred))
print('RMSE:', sqrt((mean_squared_error(y_test_log, AB_pred))))
```

# Gradient Boosting Decision Tree Regressor

Gradient boosting is similar to the Ada boost, however instead of changing the weights at every instance this method tries to fit the new predictor to the residual errors made by the previous predictor. This model again uses a decision tree as its base model.

```python
GB_model = GradientBoostingRegressor()

params = {'n_estimators':[100],
          'max_depth':[4,10,None], 
          'max_features':[1.0,0.1,'auto'],
         'n_iter_no_change':[6]} 

GS_GB = GridSearchCV(GB_model, params,cv=5,verbose=2, return_train_score=True, n_jobs=4)


GS_GB.fit(X_train_log,y_train_log)

GB_pred = GS_GB.predict(X_test_log)

```

```python
print("Best Parameters")
GS_GB.best_estimator_
```

```python
print("Gradient Boosting Model Score")
print("-----------------------------")
print("Train R-squared:", GS_GB.score(X_train_log,y_train_log))
#print("Cross Val Score: %.2f", cross_val_score(GS_GB, X_train_log, y_train_log, cv=5).mean())
print("Test Score:", GS_GB.score(X_test_log, y_test_log))
print('MSE:', mean_squared_error(y_test_log, GB_pred))
print('RMSE:', sqrt((mean_squared_error(y_test_log, GB_pred))))
```

```python
#fit the best params
best_est_GB = GS_GB.best_estimator_
f_GB = best_est_GB.fit(X_train_log, y_train_log)
```

```python
#top features
feature_GB = pd.DataFrame(f_GB.feature_importances_,
             columns=['importance'],
             index=X_train_log.columns
             ).sort_values(by='importance', ascending=False)

print("          Top 10 Features")
print("     ----------------------")
feature_GB.head(10)
```

# Random Forrest Regressor

With the improved results from Decision Tree based models, I decided to run a Random Forrest regressor as this generally yields an overall better model. As a Random Forrest is an ensemble of Decision Trees trained via the bagging method, which in returns creates greater tree diversity and thus a generally better model. 

```python
model_RF = RandomForestRegressor(random_state=1)

params = { 'max_depth': [5,10,None],
          'max_features': [0.1,0.5,"auto"],
 'n_estimators': [100,10]}

GS_RF = GridSearchCV(model_RF, params,cv=5,verbose=2,n_jobs=3)
GS_RF.fit(X_train_log, y_train_log)
```

```python
print("Best Parameters")
GS_RF.best_estimator_
```

```python
y_hat_RF = GS_RF.predict(X_test_log)

print("Random Forrest Model Score")
print("-----------------------------")
print("Train Score:", GS_RF.score(X_train_log, y_train_log)) # R2 score
#print(cross_val_score(GS_RF, X_train_log, y_train_log, cv=5).mean())
print("Test Score:",GS_RF.score(X_test_log, y_test_log))

print('MSE:', mean_squared_error(y_test_log, y_hat_RF))
print('RMSE:', sqrt((mean_squared_error(y_test_log, y_hat_RF))))
```

```python
#fit the best params
best_est_RF = GS_RF.best_estimator_
f_RF = best_est_RF.fit(X_train_log, y_train_log)
```

```python
#top features
feature_RF = pd.DataFrame(f_RF.feature_importances_,
             columns=['importance'],
             index=X_train_log.columns
             ).sort_values(by='importance', ascending=False)

print("          Top 10 Features")
print("     ----------------------")
feature_RF.head(10)
```
